```python
"""
ASSIGNMENT:1. Load a dataset simulating airport and flight data - adsb.json, oag.json files.
General functions to download data from URL and load it into a Spark DataFrame.
Main function is load_data in steps with error profing and verification.

Args: URL 
Return: Spark DF 

"""

def load_data(url):
    
    
    # Step 1: Extract the file name from the URL
    file_name = os.path.basename(urlparse(url).path)
    
    # Step 2: Download the data
    data = stream_download(url)
    if data is None:
        return None  

    # Step 3: Analyze the structure (optional)
    analyze_json_structure(data)
    
    # Step 4: Try loading into Spark without saving to disk
    try:
        df = load_spark_dataframe_from_memory(data)
        print("#4# DataFrame loaded successfully from memory #4#")
    except Exception as e:
        print(f"#4# Error loading DataFrame from memory: {str(e)} #4#")
        print("#4# Attempting to load DataFrame using file-based approach #4#")

        # Step 4.1: Fallback: Save to a temporary file and read into Spark
        tmp_file_path = save_to_temp_file(data, file_name)
        if tmp_file_path is None:
            return None

        # Step 4.2: Validate and load the DataFrame from the file
        df = load_spark_dataframe(tmp_file_path)
        if df is None:
            return None  
    
    # Step 5: Show the DataFrame
    print("#5# DataFrame 1st line #5#")
    df.show(1)
    return df

def stream_download(url):
    """Download data from the URL and return it as a string (JSON)."""
    try:
        response = requests.get(url, stream=True)
        if response.status_code == 200:
            print(f"#1# Data from {url} successfully loaded #1#")
            return response.text  # Return the raw JSON as string 
        else:
            print(f"#1# Failed to load {url}. Status code: {response.status_code} #1#")
            return None
    except requests.RequestException as e:
        print(f"#1# Error while downloading data from {url}: {str(e)} #1#")
        return None

def analyze_json_structure(data):
    """Analyze the structure of the JSON data and print a sample."""
    try:
        json_data = json.loads(data)
        if isinstance(json_data, dict):
            keys = list(json_data.keys())
            print(f"#2# Top-level keys found: {keys} #2#")
        elif isinstance(json_data, list):
            print(f"#2# Sample data from the list: {json.dumps(json_data[:1], indent=4)} #2#")
        else:
            print("#2# No valid data structure found. #2#")
    except json.JSONDecodeError:
        print("#2# Failed to parse JSON structure. Skipping analysis. #2#")

def save_to_temp_file(data, file_name):
    """Save the JSON string to a temporary file and return the file path."""
    try:
        tmp_file_path = os.path.join(tempfile.gettempdir(), file_name)
        with open(tmp_file_path, "w") as f:
            f.write(data)
        print(f"#3# Temporary file {file_name} saved #3#")
        return tmp_file_path
    except Exception as e:
        print(f"#3# Error saving temporary file {file_name}: {str(e)} #3#")
        return None

def load_spark_dataframe_from_memory(data):
    """Load the JSON data into a Spark DataFrame from an in-memory string."""
    try:
        json_rdd = spark.sparkContext.parallelize([data])
        df = spark.read.json(json_rdd)
        return df
    except Exception as e:
        print(f"#4# Error loading DataFrame from memory: {str(e)} #4#")
        raise

def load_spark_dataframe(file_path):
    """Load the JSON data from a file into a Spark DataFrame with error handling."""
    try:
        df = spark.read.json(file_path)
        print("#4# DataFrame loaded successfully from file #4#")
        return df
    except Exception as e:
        print(f"#4# Error loading DataFrame from file: {str(e)} #4#")
        print("#4# Attempting to load DataFrame using multiline option #4#")
        
        try:
            df = spark.read.option("multiline", "true").json(file_path)
            print("#4# DataFrame loaded successfully in multiline option #4#")
            return df
        except Exception as e:
            print(f"#4# Error loading DataFrame in multiline option: {str(e)} #4#")
            return None
```

```python
"""
ASSIGNMENT: 2. Use Apache Spark to ingest and process the data (e.g., data cleaning, aggregation, transformation).

Recursively flattens a DF. By explode_outer function. Original shape can be reconstructed according columns names if needed.
Alternative is multi index DF. 
Args: DF, (for oag_df, oag_multi_df)
prefix: Can stay empty or def.by user.
Returns: A flattened DF with  0 nesting. Columns names are created base on nesting depth.
"""

def flatten_json(df):
    
    # Use Complex Fields in Schema   
    complex_fields = dict([(field.name, field.dataType)
                           for field in df.schema.fields
                           if isinstance(field.dataType, (T.ArrayType, T.StructType))])

    counter = 1  

    while complex_fields:
        col_name = list(complex_fields.keys())[0]
        print(f"{counter}: {col_name} of type {type(complex_fields[col_name])}")
        counter += 1  

        # If StructType, convert all sub-elements to columns.
        if isinstance(complex_fields[col_name], T.StructType):
            expanded = [F.col(col_name + '.' + k).alias(col_name + '_' + k) 
                        for k in [n.name for n in complex_fields[col_name]]]
            df = df.select("*", *expanded).drop(col_name)

        # If ArrayType, add the array elements as rows us the explode function    
        elif isinstance(complex_fields[col_name], T.ArrayType):    
            df = df.withColumn(col_name, explode_outer(col_name))

        # Recompute remaining complex fields in schema       
        complex_fields = dict([(field.name, field.dataType)
                               for field in df.schema.fields
                               if isinstance(field.dataType, (T.ArrayType, T.StructType))])

    return df
```

```python
"""
ASSIGNMENT 2: Use Apache Spark to ingest and process the data (e.g., data cleaning, aggregation, transformation).

General functions for cleaning data in Spark DataFrame.
General Args: DF before KPIs up_adsb_df/up_adsb__multi_df,oag_flat_df/oag_milti_flat_df,  
        Args: transform_string_case(df, toUpper = True/Fals)
        Args: convert_unix_to_timestamp (DF, Unix Col (str), New col (str))
Return: handle_missing_values(df with replaced empty to 0 - can by adjusted)
        drop_constant_columns(df with dropped columns with same values)
        transform_string_case(df with changed string cases to_Upper true/fals)
        report_data_issues( return info about possible data issues)
        convert_unix_to_timestamp (DF with new column and transformed Unix time.)

"""

def handle_missing_values(df):
    df = df.replace("", None).replace(0, None)
    df = df.na.drop()  # Optional: Drop rows with any null values
    return df


def drop_constant_columns(df):
    for col in df.columns:
        distinct_count = df.select(F.countDistinct(col)).collect()[0][0]
        if distinct_count == 1:
            df = df.drop(col)
            print(f"Column '{col}' has constant values and was dropped.")
    return df

def transform_string_case(df, to_upper=True):
    for col, dtype in df.dtypes:
        if dtype == 'string':
            if to_upper:
                df = df.withColumn(col, F.upper(F.col(col)))
            else:
                df = df.withColumn(col, F.lower(F.col(col)))
    return df


def report_data_issues(df: DataFrame):
   
    print("# Data Issues Report #")

    for col in df.columns:
        issues = []  
        
        # mixed types (strings and numbers)
        mixed_type_count = df.filter(F.col(col).cast(DoubleType()).isNotNull()).count()
        total_count = df.count()
        
        if mixed_type_count > 0 and mixed_type_count < total_count:
            issues.append("contains mixed types (strings and numbers)")
        
        # null values
        null_count = df.filter(F.col(col).isNull()).count()
        if null_count > 0:
            issues.append(f"has {null_count} null values")

        # NaN values only for numerical columns
        if isinstance(df.schema[col].dataType, (DoubleType, FloatType)):
            nan_count = df.filter(F.isnan(F.col(col))).count()
            if nan_count > 0:
                issues.append(f"has {nan_count} NaN values")
        
        # zeros only for numeric columns
        if isinstance(df.schema[col].dataType, (IntegerType, LongType, DoubleType, FloatType)):
            zero_count = df.filter(F.col(col) == 0).count()
            if zero_count > 0:
                issues.append(f"has {zero_count} zeros")

        # unique values
        distinct_count = df.select(F.countDistinct(col)).collect()[0][0]
        if distinct_count == 1:
            issues.append("contains only one unique value")

        # Progress print out
        if issues:
            print(f" - {col} {', '.join(issues)}.")


def convert_unix_to_timestamp(df: DataFrame, unix_col: str, new_col: str) -> DataFrame:
  
    return df.withColumn(new_col, F.to_timestamp(F.from_unixtime(F.col(unix_col))).cast(TimestampType()))


```

```python
"""
ASSIGNMENT 6: Display results in form of table.

Display a DataFrame as a styled table using matplotlib.
Args: pandas DataFrame
Return: Table plot.
"""

def display_table(df):

    
    fig, ax = plt.subplots(figsize=(8, 4))
    ax.axis('tight')
    ax.axis('off')

    
    table = ax.table(cellText=df.values,
                     colLabels=df.columns,
                     cellLoc='center',
                     loc='center',
                     colColours=["#f5f5f5"] * len(df.columns))

    
    table.auto_set_font_size(False)
    table.set_fontsize(12)
    table.scale(1.2, 1.2)

    for (i, j), cell in table.get_celld().items():
        if i == 0:  
            cell.set_text_props(weight='bold', color='white')
            cell.set_facecolor('#494391')  
        else:
            
            cell.set_facecolor('#d9d9d9' if i % 2 else '#ffffff')

    plt.show()

```

```python
"""
ASSIGNMENT 3a: Basic KPI Airport general (all flights, on ground/off ground) average speed according:
airport of ORIGIN.
Args: DF - up_adsb_df/up_adsb__multi_df
Returns: Avg. speed per airport of origin in two columns.
NOTE 3a.1: according adsd.json 
NOTE 3a.2: static nature of data is giving same results betwen Origina and Destination airport.
"""
def avg_speed_origin(df):
    avg_speed_per_airport_origin = df.groupBy("Origin").agg(F.avg("Speed").alias("Average Speed"))
    
    return avg_speed_per_airport_origin

                                                    
```

```python
"""
ASSIGNMENT 3a: Basic KPI average speed according TYPE of aircraft in general.
Args: DF - up_adsb_df/up_adsb__multi_df
Return: Return two columns DF. 
"""
def avg_speed_type(df):
    avg_speed_per_aircraft = df.groupBy("Type").agg(F.round(F.avg("Speed"), 1).alias("Average Speed"))   
    
    return avg_speed_per_aircraft
```

```python
"""
ASSIGNMENT 3a: Basic KPI flight duration according flight of aircraft. 
Conditions Onground need to be  and Speed under 100.
Args: DF - up_df_adsb/up_df_adsb_multi
Returns: DF with two columns Flight ID and Duration (min.)
"""

def flight_duration(df):
    flight_duration = df.filter(
        (F.col("Onground") == 1) & (F.col("Speed") < 100)
    ).groupBy(F.col("Flight").alias("FlightID")) \
        .agg(
            F.min("LastUpdateTime").alias("Start Time"),
            F.max("LastUpdateTime").alias("End Time")
        ) \
        .withColumn(
            "Duration (min.)",
            F.when(
                F.col("End Time").isNotNull() & F.col("Start Time").isNotNull(),
                F.round(
                    (F.col("End Time").cast("long") - F.col("Start Time").cast("long")) / 60, 
                    2
                ).cast(DoubleType())
            ).otherwise(None)
        )
    
    return flight_duration
```

```python
"""
ASSIGNMENT 3b: Basic KPIs Total NUMBER of delayed flights DEPARTURES per Airport.

Counts the number of delayed arrivals. Count is based on *departure_actualTime_outGateTimeliness column.  
Args: oag_flat_df/oag_milti_flat_df (flattened DF).  
Returns: DF - count of delayed departures per airport.
"""

def amount_delayed_departures(df):
    delayed_departures_amo = df.filter(
        F.col("data_statusDetails_departure_actualTime_outGateTimeliness") == "Delayed").groupBy(
        F.col("data_statusDetails_departure_airport_iata").alias("Airport IATA")).agg(
        F.count("*").alias("Number of Delayed Departures"))
    
    return delayed_departures_amo

```

```python
"""
ASSIGNMENT 3b: Basic KPIs Total NUMBER of delayed flights ARRIVALS per Airport.

Counts the number of delayed arrivals. Count is based on *arrival_actualTime_inGateTimeliness column.  
Args: oag_flat_df/oag_milti_flat_df (flattened DF).  
Returns: DF - count of delayed arrivals per airport.
"""

def amount_delayed_arrivals(df):
    delayed_arrivals_amo = df.filter(
        F.col("data_statusDetails_arrival_actualTime_inGateTimeliness") == "Delayed").groupBy(
        F.col("data_statusDetails_arrival_airport_iata").alias("Airport IATA")).agg(
        F.count("*").alias("Number of Delayed Arrivals")
    )
    
    return delayed_arrivals_amo
  
```

```python
"""
ASSIGNMENT 3b: Basic KPIs Total TIME of delayed flights DEPARTURES per Airport.

Calculates the total minutes of delay for departures. (HH:MM:SS format to minutes)
Args: oag_flat_df/oag_milti_flat_df (flattened DF).
Returns: DF - total delay in minutes for arrivals per airport.
"""
def total_departure_delay_minutes(df):
    
    departure_delay = df.filter(F.col("data_statusDetails_departure_actualTime_outGateVariation").isNotNull()) \
        .withColumn("departure_delay_minutes", 
                    F.split(F.col("data_statusDetails_departure_actualTime_outGateVariation"), ":")[0].cast("int") * 60 +
                    F.split(F.col("data_statusDetails_departure_actualTime_outGateVariation"), ":")[1].cast("int")) \
        .groupBy(F.col("data_statusDetails_arrival_airport_iata").alias("Airport IATA")) \
        .agg(F.sum("departure_delay_minutes").alias("Total Departure Delay (min.)"))
    
    return departure_delay
```

```python
display_table(total_departure_delay_minutes(oag_flat_df).toPandas()) 
```

```python
"""
ASSIGNMENT 3b: Basic KPIs Total TIME of delayed flights ARRIVALS per Airport.

Calculates the total minutes of delay for arrivals. (HH:MM:SS format to minutes)
Args: oag_flat_df/oag_multi_flat_df (flattened DF).
Returns: DF - total delay in minutes for arrivals per airport.
 
"""
def total_arrival_delay_minutes(df):
    

    arrival_delay = df.filter(F.col("data_statusDetails_arrival_actualTime_inGateVariation").isNotNull()) \
        .withColumn("arrival_delay_minutes", 
                    F.split(F.col("data_statusDetails_arrival_actualTime_inGateVariation"), ":")[0].cast("int") * 60 +
                    F.split(F.col("data_statusDetails_arrival_actualTime_inGateVariation"), ":")[1].cast("int")) \
        .groupBy(F.col("data_statusDetails_arrival_airport_iata").alias("Airport IATA")) \
        .agg(F.sum("arrival_delay_minutes").alias("Total Arrival Delay (min.)"))
   

    return arrival_delay
```

```python
"""
ASSIGNMENT 4a: a. Filter the DataFrame to retain only the most recent entry (the one with the smallest LastUpdate ) for each FlightId .
NOTE: since asigment is asking most recent entry with smallest LastUpdate resulting answe would be first LastUpdate value for flight.
Args: up_adsb_df/up_adsb_multi_df DF with updated unix time.
Returns: DF - return earliest update of LastUpdateTime per flight.
 
"""
def get_earliest_flight_entries(df):
    window_spec = Window.partitionBy("Flight")
    return df.select(
        F.col("Flight").alias("FlightId"),
        F.min("LastUpdateTime").over(window_spec).alias("Earliest Update")
    ).distinct()
    
```

```python
display_table(get_earliest_flight_entries(up_adsb_df).toPandas()) 
```

```python
"""
ASSIGNMENT 4b: b. Return a DataFrame containing only the FlightId and the corresponding latest LastUpdate .
Args: up_adsb_df/up_adsb_multi_df DF with updated unix time.
Returns: DF - return latest update of LastUpdateTime per flight. 
"""
def latest_flight_entries(df):
    window_spec = Window.partitionBy("Flight")
    return df.select(
        F.col("Flight").alias("FlightId"),
        F.max("LastUpdateTime").over(window_spec).alias("Latest Update")
    ).distinct()

```

```python
display_table(latest_flight_entries(up_adsb_df).toPandas()) 
```
